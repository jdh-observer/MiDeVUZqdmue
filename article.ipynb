## Introduction

Generative artificial intelligence (GenAI) models challenge researchers in the humanities and call on them to rethink their methods, subjects and critical stance (<cite id="h3i4e"><a href="#zotero%7C22732367%2FCYYXJYJE">(Clavert &#38; Muller, 2024)</a></cite>). Historians are invited to question the uses and implications of these technologies in the production of knowledge about the past (<cite id="o1gsd"><a href="#zotero%7C22732367%2F5XVQRN76">(Meadows &#38; Sternfeld, 2023)</a></cite>). The widespread use of GenAI tools by pupils and students in their education also introduces this tension into the field of teacher training. These issues invite those involved in tertiary education to consider how to encourage the development of a critical stance among their students  and, in our case, future teachers regarding GenAI.


In this context, the history of education, as a critical and reflective discipline, appears to be a lever for understanding, contextualising and questioning the use of GenAI in schools and universities. Rooted in a tradition that is at once socio-historical, pedagogical and praxeological (<cite id="gdx2e"><a href="#zotero%7C22732367%2FBWHG9J9J">(Wagnon, 2018)</a></cite>), the history of education helps to foster a critical professional culture. It thus enables the historicisation of pedagogical reflection (<cite id="ss53n"><a href="#zotero%7C22732367%2FKK3QZMMW">(Nóvoa, 1997)</a></cite>) while providing a valuable tool to help future teachers develop an ethical, reflective and contextualised approach (<cite id="f6nga"><a href="#zotero%7C22732367%2FCFS4GG63">(Bandini, 2019)</a></cite>). Furthermore, (<cite id="5o7z5"><a href="#zotero%7C22732367%2FUBWYKEDN">(Petit, 2022)</a></cite>) emphasised the importance of not approaching GenAI as a radical break in the history of educational technology, but rather as part of a long genealogy of pedagogical transformations. This historical perspective would enable students to adopt a critical stance on the tensions at play between technical and pedagogical innovations (<cite id="h9r4f"><a href="#zotero%7C22732367%2FQVUZMCWX">(Tricot, 2017)</a></cite>).

In response to this situation, we are designing and testing a training programme that combines a historical approach (source analysis, contextualisation, critical reading and interpretation) with the use of GenAI. This programme should encourage the emergence of constructive doubt among students (<cite id="wraoi"><a href="#zotero%7C22732367%2FUBWYKEDN">(Petit, 2022)</a></cite>). The historical approach should make it possible to question the reconstructions of the past proposed by GenAI, as well as their production processes and biases (<cite id="hd59u"><a href="#zotero%7C22732367%2F5XVQRN76">(Meadows &#38; Sternfeld, 2023)</a></cite>).

After discussing the potential of adopting a historical approach (and the tools of the historian) to question the outputs of GenAI, we will outline the context, theoretical framework and methodology used to carry out this research. We will then move on to presenting and discussing the results obtained.

## Problem

Our exploratory research approach is based on the tensions between the functioning of GenAI models and the principles of historical criticism. GenAI models produce texts based on massive documentary databases. However, they operate without clearly explaining their sources, without contextualisation, and according to a probabilistic logic that does not meet the requirements of truth. Their functioning therefore consists of estimates based on reconfigured and thus "blurred" data (<cite id="5brox"><a href="#zotero%7C22732367%2FXS73FIEF">(Sternfeld, 2023)</a></cite>, <cite id="4w6rd"><a href="#zotero%7C22732367%2FUBWYKEDN">(Petit, 2022)</a></cite>, <cite id="yu7kh"><a href="#zotero%7C22732367%2FC5HCLXK7">(Kansteiner, 2022)</a></cite>).

In the field of history, this dynamic comes into direct conflict with the principles of historical criticism: gathering and evaluating sources, cross-checking documents, contextualising statements, and identifying biases. GenAI generates narratives of the past without historiography, without a critical framework, without debate – all elements that constitute the epistemology of history (<cite id="cdzfi"><a href="#zotero%7C22732367%2FFKEMVZ89">(Prost, 2010)</a></cite>).

In teacher training, the history of education can play a dual role: on the one hand, it sheds light on the historical dynamics of the production and circulation of pedagogical knowledge; on the other hand, it contributes to the development of a critical stance towards GenAI and its uses in training. To do this, students must be guided in integrating the tools of the historian. First, summarise and articulate the nature of the historical process said as "source-reading-writing" (<cite id="txx2d"><a href="#zotero%7C22732367%2FCYYXJYJE">(Clavert &#38; Muller, 2024)</a></cite>). Second, collect and evaluate sources, cross-checking documents, contextualise statements, and identify biases. Thirdly, they must analyse sources, place them in their context of production, and proceed to a critical reading and interpretation (<cite id="qkzx3"><a href="#zotero%7C22732367%2FFKEMVZ89">(Prost, 2010)</a></cite>). This approach teaches students that truth is the result of a construction of relationships and cannot be reduced to a single proposition (<cite id="4dbv4"><a href="#zotero%7C22732367%2FCHWBKL7F">(Munn et al., 2024)</a></cite>). The reconstruction of truth articulates tertiary, secondary and primary retentions that can be questioned (<cite id="6ram7"><a href="#zotero%7C22732367%2FNWCVZT6F">(Rouvroy &#38; Stiegler, 2015)</a></cite>). The use of GenAI questions the object of truth (<cite id="ryfnw"><a href="#zotero%7C22732367%2FANICGCNA">(Rovira Martorell et al., 2025)</a></cite>), and even the very notion of truth, by distinguishing between a coherentist approach to truth (specific to LLM) and an approach that articulates coherentism and correspondence (specific to humans) (<cite id="gec6b"><a href="#zotero%7C22732367%2FI8WFU6Y4">(Compagno, 2025)</a></cite>).

We therefore hypothesise that the teaching of the history of education is the place to question the use of GenAI in terms of how it works, the knowledge being constructed in a training process, and teaching practices. In other words, we believe that reflection on the "gestures of the historian" plays a role in teacher training, but that it also allows us to question the use of GenAI and professional practice. To test our hypothesis, we have focused part of a seminar in our training programme on the history of education approach. To stimulate questions among students and encourage debates that could link the functioning of GenAI, historical approaches and professional practices, we designed activities dedicated to the figure of Henri Roorda
and the history of certain educational approaches.

## Theoretical framework

To design the training programme and analyse the practices it induces, we refer to the philosophy of technology (<cite id="jfwki"><a href="#zotero%7C22732367%2F4FNYMB73">(Alombert, 2025)</a></cite>, <cite id="t77yb"><a href="#zotero%7C22732367%2FDAIV7H5X">(Stiegler, 2020)</a></cite>) and the historical approach (<cite id="5ha2p"><a href="#zotero%7C22732367%2FB3HYYQZY">(Bloch, 1949)</a></cite>, <cite id="cytkn"><a href="#zotero%7C22732367%2FFKEMVZ89">(Prost, 2010)</a></cite>). We specifically focus on the possibility of problematising, in training, the relationship between the functioning of GenAI and the historical approach to question how truth – or more modestly, our frames of reference – is constructed regarding educational practices.

We therefore try to encourage students and future teachers to adopt an attitude that allows them to avoid delegating their thinking to GenAI (<cite id="ogt3n"><a href="#zotero%7C22732367%2FKCEBK8KZ">(Perrin et al., 2025)</a></cite>). Technology is an integral and constitutive part of humanity. Technology is a Pharmakon (<cite id="vk493"><a href="#zotero%7C22732367%2F4FNYMB73">(Alombert, 2025)</a></cite>): it is neither good nor bad in itself. The challenge is therefore to think with GenAI, to use it as an aid rather than a prosthesis (<cite id="v0b18"><a href="#zotero%7C22732367%2FKCEBK8KZ">(Perrin et al., 2025)</a></cite>).

The challenge for students is therefore to learn to identify and question the categories induced using GenAI and, consequently, the categories of thought (<cite id="fg93j"><a href="#zotero%7C22732367%2FRP8HL7NK">(Perrin &#38; Piot, 2024)</a></cite>), in the historical approach, in training practices and in professional action.

However, such an approach is paradoxical, as it would require students, who are not yet experts in the content covered in seminars, to question the quality of texts generated by AI (<cite id="bnnoq"><a href="#zotero%7C22732367%2FKCEBK8KZ">(Perrin et al., 2025)</a></cite>). This is why the role of the collective is essential in questioning the categories induced by technology by conducting controversies, which reflect a formulation of thought held by argumentative constraints (<cite id="sk239"><a href="#zotero%7C22732367%2FBTP6BM2N">(Alombert, 2024)</a></cite>). Controversies concerning the functioning of GenAI and the historical approach should enable the development of critical thinking. (<cite id="tjh4i"><a href="#zotero%7C22732367%2F8337629H">(Boisvert, 2000)</a></cite>, <cite id="lecos"><a href="#zotero%7C22732367%2FJQCJ539L">(Boisvert, 2015)</a></cite>) highlights the need to specify how to operationalise the critical thinking expected of learners. He emphasises the importance of distinguishing between abilities (specific cognitive skills for analysing, evaluating or reasoning), attitudes (ways of acting, such as open-mindedness, intellectual perseverance or a desire to be well informed) and strategies (ways of organising and regulating action such as comparing similar situations, developing criteria for evaluation or planning one's analytical approach). He also specifies that critical thinking requires knowledge of the subject under examination.

In our case, this involves contextualising the two themes covered. The aim of this approach is to encourage students to question how the technique deployed by GenAI works and what actions are required in terms of historical, educational and professional approaches. Students will therefore examine how the GenAI works when it comes to reconstructing historical thinking or educational approaches and will put the GenAI outputs into perspective using a historical approach.

## Method

### A design approach

A Design Based Research approach (<cite id="jatvp"><a href="#zotero%7C22732367%2FESFQP82G">(Barab &#38; Squire, 2004)</a></cite>) enables us to design, implement and improve our training programme. This method enables us to clarify and test our design hypotheses, as well as to design and improve our training programme in an iterative process (<cite id="xd3t9"><a href="#zotero%7C22732367%2FKCEBK8KZ">(Perrin et al., 2025)</a></cite>). As part of this approach, we therefore draw up design hypotheses that guide the design of our training programme. The implementation of all or part of this programme then enables us to test and revise our design hypotheses, or the way we operationalise them in the training programme.

To operationalise our approach and set up our exploratory research, we construct our design hypotheses with reference to (<cite id="9sl2h"><a href="#zotero%7C22732367%2FR6EN794Y">(Knibbe, 2016)</a></cite>). He proposes three parts: a) understanding or describing the phenomenon that we need to consider, namely the (potentially) instrumented activity of students, b) the characteristics of the programmes that we can influence to guide the instrumented activity of students, and c) the expected effect on the instrumented activity of students. We therefore formulate our hypotheses as follows: a) given such phenomenon(s), b) by acting on such characteristic(s), c) we hope to obtain such effect(s).

### Design hypotheses

As part of this design-oriented research, we are testing the following hypotheses:

**Hypothesis 1:** a) Given the paradox for students of having to question the relevance of the generated texts while not being experts in the field, and given that a comparative approach allowing them to examine similarities and differences is one way of addressing this paradox, b) the training programme aims to compare the results obtained using AI, the historical approach and the functioning of GenAI c) in order to promote the development of critical thinking, which consists of identifying different results and hypothesising about the origin of these differences.

**Hypothesis 2:** a) Given that the history of education helps to foster a critical professional culture conducive to the analysis of continuities and discontinuities, b) if students write biographies of Roorda or trace the history of certain educational systems using GenAI, then compare these texts with scientific works and the historical approach (under the guidance of a specialist and through group interaction), c) then they will develop the ability to question and nuance the validity of the discourse generated by GenAI, which should lead them to recognise the limitations of using GenAI for historical reconstructions (but not only) and to adopt a more critical and reflective stance towards these tools.

**Hypothesis 3:** a) Given that controversy promotes the ability to question the categories induced by technology (whether those of GenAI or those that formalise professional practice), b) controversies concerning the nature of texts generated by GenAI, when viewed in context and linking the functioning of GenAI to historical, pedagogical and professional practices, c) promote critical thinking that questions what enables us to judge the validity of the recommendations on professional practices proposed by the generated texts, doxas  and prescriptive texts.

**Hypothesis 4:** a) Given that it is important to understand the difference between a coherentist approach to truth (specific to LLMs) and an approach to truth that articulates coherentism and correspondence (specific to humans); that the truth of LLMs is not explainable and that the truth of humans always depends on the perceiving subject; that truth is therefore a temporarily stabilised social construct that is important to understand in order to accept it, b) the confrontation concerning the "interpretation" of a "source" generated by a GenAI by a collective adopting a historical approach and questioning professional ethics, c) promotes critical thinking that accepts questioning a product (generated text) in relation to relevant truth regimes.

Across these hypotheses, several outcomes are expected. After the teaching sequences, students will be able to identify the limitations, errors or simplifications present in texts generated by GenAI; students will draw on scientific sources and elements of the historical method to support their critiques (comparison and analysis of primary sources, contextualisation of these sources using existing secondary literature on the subject). In the longer term, the following outcomes could also be expected: students will develop a broader critical and reflective stance towards AI-generated outputs, which will also enable them to question the concepts of truth and interpretation (by putting them into perspective with the concept of probability); students will recognise the specificities and value of the historical approach in relation to automatic reconstructions; students will change their habits regarding the use of GenAI.

### A training program

This research-oriented project involves designing, implementing and evaluating two sequences that form part of an advanced seminar offered as part of initial training (Bachelor's degree in primary education, Haute Ecole Pédagogique Vaud) for future primary school teachers in the canton of Vaud. This advanced seminar (BP53Apy) aims to take a critical and in-depth look at active teaching methods, their history and their use in classrooms today. Interest in these teaching methods today is justified by their role in building inclusive and learning-friendly school environments, as well as their continuing influence on contemporary teaching practices (<cite id="7pljn"><a href="#zotero%7C22732367%2FU4BRTGRN">(Grootaers &#38; Tilman, 2009)</a></cite>, 
<cite id="8y2tj"><a href="#zotero%7C22732367%2FTLCPIA2L">(Vellas, 2022)</a></cite>,
<cite id="wd2vr"><a href="#zotero%7C22732367%2FKNJXZZWY">(Wagnon, 2018)</a></cite>,
<cite id="462qv"><a href="#zotero%7C22732367%2FYUAVNDD7">(Wagnon, 2021)</a></cite>).

This seminar has several objectives, namely: to deepen knowledge of active teaching methods and their origins by adopting a historical approach; to design a project that includes an active teaching method; to project oneself into one's future classroom by envisaging a teaching sequence that includes active teaching.

This teaching framework seemed to us to be a particularly relevant field in which to introduce a reflection on the uses of GenAI in training, and more specifically on the relationship between history and GenAI. The seminar provides a space conducive to questioning, in a collective and participatory approach to the analysis of practices, how these tools redefine the relationship to knowledge and raise the question of the concept of truth in relation to texts produced by GenAI.

### Teaching sequences

Two three-hour teaching sequences were designed and implemented within a module taught over one semester at a rate of three hours per week (Appendix 2). The first was designed before the start of the module. The second was designed to complement the first. The design assumptions were reviewed and refined by the research team between the first and second sequences.

The first sequence took place in five stages: 1) writing a biography of Roorda using GenAI, 2) group discussion led by a specialist historian, 3) group comparison of the biographies generated by the GenAI and the one reconstructed using the historian's method, 4) implementation of a historian's approach to clarify Roorda's thinking, 5) objectification of the uses of GenAI to explain and promote the transfer of critical thinking.

The second sequence took place in three stages: 1) development of a descriptive sheet for an active teaching method using GenAI and secondary literature, 2) implementation of a historical approach to trace the history of these methods and thus contrast the production and uses of GenAI in this context, 3) objectification of the uses of GenAI to explain and promote the transfer of critical thinking.

### Participants

This research was conducted during the autumn semester of 2025 at HEP Vaud. It was carried out by an interdisciplinary team active in the fields of educational history, educational engineering and technology sciences.

The audience for this research consists of a group of nine students (including two men, with an average age of 23) enrolled in this training seminar. The participants in this study reported that they regularly use GenAI tools, with 62% of them using them specifically for training purposes. In this context, GenAI tools are used to better understand or explore topics in greater depth (50%) or as a tool to save time (38%). The GenAI tool most used by our audience is ChatGPT. Most of the people involved use these GenAI tools several times a week. However, they do not consider themselves to be very proficient in GenAI tools, and their training in GenAI tools has mainly been self-taught or with the help of peers.

### Data construction and analysis

#### Participant observation

A participant observation approach was conducted by one of the authors involved in the training. This two-step approach (<cite id="khy55"><a href="#zotero%7C22732367%2FGWXQ9QKI">(Beaud &#38; Weber, 2003)</a></cite>) consisted first of taking notes after each training session, adopting three complementary attitudes: a) avoiding preconceived tought through a game of proximity/distance, b) being sensitive to the needs of the students and the nature of the support offered by focusing on the variability of their actions, and c) establishing a horizon of expectation by recalling the training objectives and key points of the design hypotheses before the training, then setting them aside to welcome the unexpected. In a second stage, the notes were reread and supplemented based on the design assumptions. This approach made it possible to re-immerse oneself in the session, recall activity configurations, describe relevant elements and outline analysis trails.

This participant observation is part of a research-training approach that combines three roles: trainer, designer and researcher (<cite id="b4ggf"><a href="#zotero%7C22732367%2F2SLK5KN3">(Perrin et al., 2022)</a></cite>). The articulation of these three positions was made possible by drawing on the collective of trainers involved in the training to guide educational decisions, considering the ecology of the training, and on the collective of researchers to achieve the best compromises between the multiple constraints associated with the research-training approach. The articulation of these three positions made it possible to design the training programme in a way that was consistent with its use (<cite id="4ewra"><a href="#zotero%7C22732367%2FLAZ5TY7R">(Béguin &#38; Cerf, 2004)</a></cite>).

#### Questionnaire

In addition, a questionnaire (Appendix 3) was designed to better understand what was significant for students in terms of integrating GenAI into a historical approach to critical thinking training. The aim was to better understand how the two training sequences influenced the relationship with GenAI, in terms of usage, representations and the ability to exercise constructive doubt when faced with the writings produced during the training.

This questionnaire is based on an operationalisation table (Appendix 3) that explains the dimensions and indicators of critical thinking, namely the abilities, attitudes and strategies involved in critical thinking that questions the use of GenAI in the historical approach. It also includes a question allowing students to describe a significant situation from the seminar that enabled them to answer the other questions in the questionnaire. This questionnaire was completed by 8 out of 9 participants.

Data analysis consisted of identifying correlations between responses to closed questions and content analysis (<cite id="9focj"><a href="#zotero%7C22732367%2F3ESL5937">(Bardin, 2001)</a></cite>) of responses to open questions. This made it possible to test the validity of the design hypotheses.

## Results

We will organise the presentation of our results in three stages. These three stages correspond to the two teaching sequences, with a third stage dedicated to questioning the potential effects of our sequences on students' practices regarding the use of GenAI.

As a reminder, during the first teaching sequence (03.10.2025), our objective was to confront students with constructive doubt. To do this, we first invited the group to use GenAI to reconstruct the biography of Henri Roorda. We then compared these reconstructions with a biography of the figure based on a cross-referenced study of several historical sources. To measure the impact of this activity on their critical thinking and reflexivity, we scheduled a discussion session, which was recorded and transcribed. Personal notes taken by one of the trainers involved in the sequence were used to supplement this data.

Analysis of the verbatim transcripts of these exchanges revealed the following:

Firstly, several students raised criticisms of the GenAI output. These criticisms included the inaccuracy, generalisation and incompleteness of the biographical reconstruction proposed by the GenAI. The portrait generated was described as "partially complete" (Student 3).

Secondly, some students pointed out a difference between the way the GenAI carried out this reconstruction and the way it might have been done by someone who, like us, adopts a proven historical approach. In this regard, criticisms were raised about the GenAI ability to summarise, notably by stating that the GenAI "sometimes reads something and summarises it in a way that is not entirely how we would understand or summarise it" (Student 1). The inability of GenAI to offer " 'interference' with the text, thus removing 'the human side' " (Student 2) from the understanding of texts and their perspective, seems to be perceived as problematic by students. This highlights the tension between the coherentist approach to truth (specific to LLM) and an approach that articulates coherentism and correspondence (specific to humans) (<cite id="o81ya"><a href="#zotero%7C22732367%2FI8WFU6Y4">(Compagno, 2025)</a></cite>) mentioned in the problem statement.

Thirdly, although some students highlighted the benefits of using GenAI for this exercise, particularly in terms of saving time when conducting initial documentary research ("chat, give us a few of the sites he read, and then we can also go back to them, read what he wrote, and then form our own opinion", Student 1) and thus to facilitate an initial sorting of information ("it already gives us more information that we can process ourselves", Student 1), they also emphasised the importance of knowing the limitations of the tools deployed. As exemplified by one student, in the exercise in question, the tensions that run through Henri Roorda's biography do not appear in the biographical reconstructions carried out with GenAI. It is therefore necessary, according to the students, to adopt a critical eye, because "if we don't read between the lines of what it offers, we don't realise everything that was at stake behind it" (Student 2).

During the discussion, these criticisms led the students to identify the limitations of using GenAI, which do not allow for reliable historical reconstruction. To address these limitations, the students raised the need to engage in collective reflection ("and so, it's true that I think it has limitations and it's important to talk about it because we don't talk about it enough here. And here we have concrete proof of what we have experienced", Student 2). They also mentioned the importance of being trained to identify these limitations ("And then you have to read between the lines, but if you've never been trained to obtain information other than through AI, you'll never have that information", Student 3). Finally, they also suggested ways to counter these limitations. These are part of a historical approach of criticism and comparison of sources ("So that also really explains why we need to look for sources other than just generators", Student 3).

Analysis of the discussions shows that some of the students seem able to identify the limitations of GenAI (particularly in terms of inaccuracy, generalisation and lack of depth in the comments generated). Some students also seem to have compared the GenAI production process to the historical approach. They criticised GenAI productions and found ways to counter them by returning to the historical approach, in particular by emphasising the importance of comparing sources. Finally, some of the students recognised the need for better training and education to be able to identify the problematic elements that characterise the content generated by GenAI. This first sequence thus seems to confirm our initial design hypothesis: the comparative experience seems to have encouraged (or facilitated) the emergence of a critical stance towards GenAI.

Our sequence, and particularly the support provided in adopting a historical approach, seems to have encouraged students' ability to nuance the validity of the content offered by GenAI (in this case, the reconstruction of Henri Roorda's biography), which supports our second design hypothesis. This confrontation also enabled students to highlight the generalisation of the statements made by the GenAI. It prompted them to use inductive categories of analysis to examine the GenAI output (Perrin and Piot, 2024), a dynamic that we mentioned in our third hypothesis. These same elements allowed students to grapple with the complexity of historical reconstruction, and in particular the importance of contextualising and comparing sources to embark on a quest for historical truth, which allows us to make a connection with our fourth hypothesis.

Although the discussion was quite dynamic, as indicated in the trainer's notes, only two-thirds of the students participated actively. The attitude of the students who did not participate particularly led the trainer to question the relevance of the activity and the possibility of repeating the experience to involve the whole group in the discussion. These post-sequence reflections were discussed with the project group, which plans to set up a second sequence with the aim of creating a new opportunity for reflection on the issues surrounding the use of GenAI.

As a reminder, the second sequence (10.10.2025), like the first, aims to confront students with constructive doubt. To do this, we invited the seminar group to use GenAI to trace the origins of a series of active teaching methods, such as field trips, free writing, workshops and work plans. In order to encourage a critical approach to the outputs of GenAI, while providing a clear framework for the activity, we gave the students the following instructions for creating descriptive sheets for these tools: "Use generative artificial intelligence (GenAI) to help you describe, document and analyse this tool, while taking a critical look at the information produced" while clarifying the objectives ("the aim of this activity is to understand the tool and assess the relevance of the contributions made by the GenAI used"). We then invited the students to compare the GenAI outputs with scientific studies on the concerned pedagogical tools under investigation. As in the first sequence, to measure the impact of this activity on their critical thinking and reflexivity, we scheduled a discussion session, which was recorded and transcribed. Here too, personal notes taken by one of the trainers involved in the sequence supplemented this data.

Analysis of the verbatim transcripts of these exchanges revealed the following points.

As in the first sequence, the students raised criticisms of the GenAI work. Among these criticisms we found again those concerning the imprecision and generalisations and the lack of contextualisation in the responses produced by the GenAI ("To say that 'workshops are widely used in primary education' is true today, but it is a generalisation: it depends greatly on the context, the teachers and the country", Student 4).

During this discussion, the students once again highlighted the difference between an approach based on the analysis and comparison of different sources, on the one hand, and the way GenAI work, on the other. They highlighted the limitations of the responses generated by GenAI ("it will sequence the different sources, which aren't really sources, but just concepts it has found on the Internet. So that's a bit... In fact, it's very enumerative and very point by point", Student 2). This time, these elements are addressed by the students in an even more incisive manner, leading them to debate the issue more deeply, as shown by this statement: "So really, the fact that it doesn't have access to scientific data like we do, when we look for articles, when we compare the two, we realise that the answers are quite limited and, as she [Student 3] said, they're very general, they don't really go into the depth of the information. So, to get a general idea, I think it's good, it's quick, we don't need to open lots of websites, it gives us a bit of an overview. But if we look for more detail, I think that even when we ask specific questions, we never really get very comprehensive, very detailed answers" (Student 5).

In this case too, the students mention the benefits of using GenAI, describing them as "time-saving" tools for performing a series of tasks such as sorting information. However, this time, these elements are nuanced by highlighting the limitations rather than the advantages of this "timesaving" strategy: "Finally, I don't think we could use ChatGPT alone to build a history of the methods we've seen today. But by comparing the information, we can still figure out what it's telling us that is correct. And what should we explore further by doing a little more scientific research? I think a mix of the two is good" (Student 5). The analysis of the verbatim transcripts therefore shows the adoption of a critical stance towards the outputs of GenAI. Indeed, the students raise a series of criticisms while identifying the errors and limitations of GenAI.

During this discussion, the students gave the impression that they had consolidated their critical stance and were becoming increasingly adept at highlighting the limitations and conditions of use of GenAI. In our opinion, the following statement shows a greater awareness than in previous discussions: "AI proceeds by rapid synthesis, linking known figures and ideas without always reflecting the complexity of scientific debates. It emphasises points of consensus (cooperation, autonomy, differentiation) but smooths over historical conflicts or controversies. It produces a linear and simplified narrative, whereas scientific texts show a history of tensions, reinterpretations and institutional contexts. In short, AI constructs a rather popularised form of knowledge: useful for introducing a subject, but insufficient for grasping historical nuances" (Student 4).

The overall analysis of this exchange points to a dynamic of progress in the development of a critical stance among students, who are demonstrating an increasingly nuanced understanding of the issues and limitations of GenAI, particularly regarding historical reconstruction. Their stance therefore seems to be increasingly reflective, particularly regarding the tension between pragmatic use (as a "timesaving" tool) and the epistemic limitations of AI-generated texts (simplifications, errors, generalisations, etc.).

This second sequence seems to confirm our first and second hypotheses. The discussion also shows that the system designed and implemented during the session in question facilitated debate, which in turn helped to consolidate the students' critical thinking. This dynamic allows us to make a connection with our fourth hypothesis.

Finally, the discourse analysis highlights the students' growing awareness of how GenAI work and their limitations in reconstructing historical truth ("we were really confronted with the reality of what ChatGPT brings us", Student 2). The highlighting of linear and simplified narratives, the omission of references and the emphasis on the complexity of the historian's method also allow us to make a connection with our fourth hypothesis.

The notes taken by one of the trainers confirm this dynamic: even though student participation remains variable (with some students participating less than others), the discussion is perceived as more impactful and the debate and arguments are perceived as more incisive, leading to a greater awareness of the issues surrounding the use of GenAI. According to the trainer, this dynamic was made possible by better support during the activity (thanks to the creation of a roadmap with explicit questions aimed at guiding reflection step by step) and also by more explicit supervision during the institutionalisation phase.

After analysing the two teaching sequences, we will attempt to examine the potential effects of our sequences on students' practices regarding the use of GenAI. To do this, we will rely primarily on the students' responses to the questionnaire.

In general, students say that the activities offered in the seminar helped them better understand how GenAI generates its responses (76%) and, above all, identify biases or limitations in GenAI outputs (100%) and develop a constructive scepticism towards the information generated (100%).

Among the activities carried out, students identified the following as most helpful in adopting a critical stance: reconstructing the biography of Henri Roorda (50%) and reconstructing the origins of a pedagogical tool derived from active teaching methods (40%). On the other hand, comparing sources is rarely mentioned (10%). These choices are also confirmed when students are asked to think of a moment during the seminar when the use of a GenAI supported their thinking and, conversely, a moment when the GenAI hindered it. The responses provided show that three out of eight respondents mention the adoption of an approach requiring the comparison of several sources as necessary to counter the limitations of AI-generated content. When students are asked what they have learned about GenAI, some mention the lack of transparency of the data used to produce the responses and the veracity of those responses.

Following our activities, some students said they now use GenAI differently and do so in a more critical and thoughtful manner (38%). Others said they use it differently without having changed their practices (25%). In both cases, what seems to have changed is mainly the fact that they now verify the answers provided by GenAI, in particular by cross-referencing several sources.

Other students do not believe they have changed the way they use GenAI (38%), explaining this stance by the fact that they already knew how they work.

Overall, both the discussions and the responses provided in the questionnaire show that the activity devoted to Henri Roorda (and the reconstruction of his biography) is a recurring theme in the exchanges, which could suggest that it was a trigger for controversy. This recurrence shows not only the potential of the historical figure of Henri Roorda to provoke reasoned debate, but also that the history of education can be seen as a fertile ground for engaging students in the analysis of epistemic tensions related to the production and use of GenAI, highlighting the historical and methodological issues that characterise reconstructions (and, more generally, responses).

## Discussion

This exploratory research opens reflection on several elements. Firstly, the system put in place has shown that the historical approach can be used as a tool for reflecting on the mechanisms of GenAI production and their limitations. In other words, the research results shown that, in the context of teacher training, the history of education provides fertile ground for collective reflection on the tensions between the probabilistic functioning of GenAI (and their production) and the requirements of the historical approach. This has enabled us to answer our first hypothesis.

Secondly, the fact that the sequence dedicated to the figure of Henri Roorda was used by students as a meaningful activity for reflecting on GenAI allowed us to emphasise that the history of education does not only play an illustrative (or even anecdotal) role, but that it is a catalyst for debate and controversy. In this regard, as the students' feedback shows, Henri Roorda's biography, marked by ideological and pedagogical tensions, among other things, lends itself well to highlighting what the reconstructions produced by the GenAI tend to smooth over (such as biographical tensions). The comparison between the narratives produced by GenAI and those reconstructed using a historical approach led students to question what had been simplified or even forgotten. Ultimately, focusing the first sequence on a historical figure such as Henri Roorda served as a catalyst: it helped to crystallise the emerging tensions between automated productions and historical reconstructions. This first sequence therefore had the merit of bringing into dialogue the epistemic issues related to reconstructions of the past and the professional and training issues related to the use of GenAI. Thus, the history of education is not only a subject to be taught, but also a lever for training reflective teachers who can question discourses, situating them and constructing a professional stance rooted in history, criticism and ethics (<cite id="boftp"><a href="#zotero%7C22732367%2FCFS4GG63">(Bandini, 2019)</a></cite>).

We can therefore assert that history, and particularly the history of education, can be understood as an epistemological means of understanding technology. Indeed, the two sequences implemented transformed GenAI into an object of study and training, elevating it beyond its simple status as a 'timesaving' tool or a means of delegating work. After our sequences, we noticed that students continued to work with GenAI, but while trying to take a more critical and reflective approach.

Our exploratory research allows us to formulate a fourth and final observation. The analysis of the sequence and of our own activity, carried out through observant participation (<cite id="92507"><a href="#zotero%7C22732367%2FGWXQ9QKI">(Beaud &#38; Weber, 2003)</a></cite>), highlighted the need to put in place conditions to encourage the emergence of controversy. In our case, this seems to have been encouraged by the trainers' supervision, the structuring of tasks and the creation of a shared space for discussion.

Our method may also be questioned. We adopted a design-based research (DBR) method, which aims to contribute both to the theorisation of training programmes and to improving their design (<cite id="j4bpe"><a href="#zotero%7C22732367%2FESFQP82G">(Barab &#38; Squire, 2004)</a></cite>). We attempted to clarify the theoretical propositions underlying the implementation profiles of our programme (<cite id="uhtgh"><a href="#zotero%7C22732367%2FHX93GTXW">(Collins, 2010)</a></cite>). To do this, we adopted a formulation of design hypotheses that articulates three propositions (<cite id="l340i"><a href="#zotero%7C22732367%2FR6EN794Y">(Knibbe, 2016)</a></cite>): a) given such phenomenon(s), b) by acting on such characteristic(s), c) we hope to obtain such effect(s).

Our approach has enabled us to clarify some elements of a local theory articulating the characteristics of our system, the cognitive activity of students and the expected effect (<cite id="c922d"><a href="#zotero%7C22732367%2F6UVHNG6V">(Gravemeijer &#38; Cobb, 2013)</a></cite>). However, it must be part of a research trajectory (<cite id="5g8qt"><a href="#zotero%7C22732367%2F3PUA8ENY">(Sandoval, 2014)</a></cite>) that allows us to structure iterations between cycles focused on a) proof of existence to verify the structure of our hypotheses, b) variations based on principle to refine implementation profiles, c) contextual variations to characterise implementation, and d) scaling to verify the learning ecology.

In this research, we began to refine the implementation profiles. In particular, we highlighted two key points concerning the construction of critical thinking. The first distinguishes between criticism and limitation of a generated text. Criticism consists of identifying problematic generated texts, while limitation involves understanding why they are problematic. The identification of a limitation stems from a comparison between the text generated by AI and the text written by a historian on the same subject. We therefore emphasise that the comparison focuses on the product and not on the process of the historian's approach. However, we must qualify our interpretation. This comparison of products is made with the actors – including historians – who participate in the process of constructing truth.

The second point concerns the relationship between the "actions of the historian", the use of GenAI and professional practice. We note the importance of the subject of historical analysis (the figure of Roorda and his reflections on classroom management) as a driver of critical thinking about the difference between text generated by AI and text written by a historian. It will be important to explain to students how a controversy arises between the past of the historical object and the present of the professional act. Our initial findings will therefore need to be corroborated by studies that articulate more precisely a) high-level conjectures (how to support the type of learning targeted), b) design conjectures (articulating the characteristics of the system and the mediations made possible by these characteristics) and c) theoretical conjectures (articulating the mediations and expected effects) (<cite id="4abpl"><a href="#zotero%7C22732367%2F3PUA8ENY">(Sandoval, 2014)</a></cite>).

Several criticisms can therefore be raised. Firstly, our research is exploratory and takes place within a specific institutional framework, namely, an elective seminar with a limited number of students. Our results cannot therefore be generalised and must be interpreted within the context described above. Secondly, the trainer observed varying levels of student participation during discussions. The development of a critical stance may therefore not have affected the whole group, or at least not in a uniform and consistent manner. Thirdly, participant observation could also be seen as a limitation, because although we took every possible precaution, this approach may introduce interpretative bias. Finally, our study aimed to measure (or at least attempt to measure) the adoption of a critical stance over the course of a single semester. It therefore does not allow us to assess the potential impact of our teaching sequences on students over time and in other contexts, particularly professional ones.

However, our research also has its strengths. Firstly, we can highlight an original link between the history of education and GenAI. Through our approach, we have established a unique dialogue between the historical approach and the critical analysis of GenAI. We have done so while exploring trails for thinking about the training of future teachers in GenAI in a context marked by a reconfiguration of access to knowledge. Secondly, to bring understanding trails to our problem, we have adopted a mixed methodological approach that requires us to step outside our respective fields of expertise and exchange perspectives. This brings us to our third and final strength: the interdisciplinary nature of our team, which brings together specialists in the history of education, philosophy, technology and engineering sciences, and in teacher training. This collaborative work forced us to clarify our respective approaches and to engage in dialogue about our usual approaches and analytical frameworks to co-construct a common language for addressing our research subject, which is also at the intersection of several disciplines. This process of clarification and constant confrontation was a powerful driver of scientific creativity, while opening original action trails to grasp the challenges of GenAI, the historical approach and professional practice.

## Appendix

### The figure of Henri Roorda

Born on 30 November 1870 in Belgium, Henri Roorda arrived in Switzerland with his family and settled in the Lake Geneva region in 1872. He lived there until his suicide on 7 November 1925. Throughout his life, he questioned the place of pupils in the school system and practices that encouraged their activity.

Influenced from an early age by his father's friends (including Pierre Kopotkine, Elisée Reclus and Ferdinand Domela Nieuwenhius), Henri Roorda built his educational philosophy on libertarian and anarchist principles <cite id="ny7rh"><a href="#zotero%7C22732367%2FU3KEUQIF">(Ausoni, 2021)</a></cite>. He wrote several books, two of which were bestsellers: Le Pédagogue n'aime pas les enfants (The Teacher Does Not Like Children, 1917) and Avant la grande Réforme de l'an 2000 (Before the Great Reform of 2000, 1925). The first established his reputation and led readers to question and criticise the learning conditions of pupils at the time. The second seems to be more forward-looking, proposing actions and reflective approaches for teaching professionals. Far from seeking to produce scholarly texts or propose a single method, he based his reflections on observations drawn from his own professional experiences as a teacher in a girls' secondary school (1892), then at the Cantonal College and Classical Gymnasium in Lausanne (1905).

Henri Roorda often compares children to defendants; "and because teachers adopt the tone and methods of judges from day one, pupils naturally take on the attitude of defendants" (<cite id="s6d07"><a href="#zotero%7C22732367%2FA4HSD3EL">(Roorda, 1926)</a></cite>), which, paradoxically, is never a central concern for schoolteachers. Teachers certainly have limited scope for action, imposed by the legislator, but they nevertheless have some room for manoeuvre (promoting the interests of pupils and learning from their mistakes, integrating "useful knowledge", intellectual integrity, etc.), and it is precisely this scope for action that Roorda draws our attention to: he does not seek to reform the school, but to teach in a different way. This translates into a desire to make children happier by motivating and surprising them, which would make them more interested and, as a result, more involved in their learning, which can only improve.

His commitment to change did not stop there. Roorda also helped to write the Declaration of Principles of the Ferrer School in Lausanne. However, no doubt for fear of losing his job, he refrained from becoming more involved. This tension between the need to be active in a system he denounced was summed up in this letter to his friend Amédée Dubois: "Two hours in the office does not prevent one from being free in one's thoughts and actions. Unless one is rich, one must think twice before giving up one's salary. Being forced to earn a living the hard way is what takes away all freedom" (letter to Amédée Dunois, quoted by Ausoni and Delacrétaz, 2025, p.57).

The fear of losing his livelihood undoubtedly contributed to his use of the pseudonym Balthazar when writing some 700 columns in various regional newspapers, some of which were very radical. This persona allowed him to express his verve, his humorous creativity and his thoughts more freely. Balthazar became such an important part of his life that he signed his last work, "Mon suicide" [1925] under both names (<cite id="96gtn"><a href="#zotero%7C22732367%2F24TYPZXS">(Ausoni &#38; Delacrétaz, 2025)</a></cite>). It should be noted that here again, Henri Roorda, ahead of his time, explored themes that are still sensitive today, such as human value, freedom to love, but above all, freedom to die.

His writings do not seem so far removed from our current concerns as teachers and educators. If we consider our questions about the presence and use of technology in education and the colossal amounts of knowledge available via the internet or GenAI, they seem to echo the "Encyclopaedic Offices where the curious would find all the information they desire" (Roorda, 1917, p.28) that Roorda described more than 100 years ago. As Alain Ausoni mentions at the end of his 2021 article, we will never stop Roordaring.
